Multithreading in python

```python
class Hello:
    def run(self):
        for i in range(5):
            print('Hello')
            
            
class Hi:
    def run(self):
        for i in range(5):
            print('Hi')
            
    
a = Hello()
b = Hi()

a.run()
b.run()
```

```python
from threading import*

class Hello(Thread):
    def run(self):
        for i in range(50):
            print('Hello')
            
            
class Hi(Thread):
    def run(self):
        for i in range(50):
            print('Hi')
            
    
a = Hello()
b = Hi()

a.start()
b.start()
```

```python
from threading import*
from time import sleep

class Hello(Thread):
    def run(self):
        for i in range(5):
            print('Hello')
            sleep(1)
            
class Hi(Thread):
    def run(self):
        for i in range(5):
            print('Hi')
            sleep(1)
    
a = Hello()
b = Hi()

a.start()
sleep(0.2)
b.start()

a.join()         #a makes main thread to wait till complete execution
b.join()         #b makes main thread to wait till complete execution

print('Bye')     #now main thread executing this statement only after a abd b threads complete their execution
```
